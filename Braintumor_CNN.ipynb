```python
#Defining image size and batch size
IMG_SIZE = 224
BATCH_SIZE =32
```

```python
#creating training data
train_datagen= ImageDataGenerator(
     rescale=1./255,
    validation_split=0.2
)
```

```python
#creating training data with above parameters
train_grenerator=train_datagen.flow_from_directory(
    r"C:\Users\PEKETI YAJNICA SAI\Downloads\Train-20240301T070009Z-001\Train",
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)
```

```python
#creating validation data with above parameters
val_grenerator=train_datagen.flow_from_directory(
    r"C:\Users\PEKETI YAJNICA SAI\Downloads\Train-20240301T070009Z-001\Train",
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

```python
#defining the model
model=keras.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
    
```
