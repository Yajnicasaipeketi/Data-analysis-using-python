```python
import pandas as pd
```

```python
mydf=pd.read_csv("/content/mydf.csv")
mydf
```

```python
import seaborn as sns
```

```python
#lineplot
pl=sns.lineplot(x="Names",y="SEM1",data=mydf)
```

```python
#relplot
sns.relplot(
    data=mydf,
    x="SEM1",y="Names",col="Placed")
#col==> Graphs are separated based on this column
```

```python
#create a relational plot using age as y axis index as x axis seperated by class
#diabetes.xlss
dfe=pd.read_excel("/content/diabetes.xlsx")
dfe['Index']=range(1,769)
sns.relplot(
    data=dfe,
    x="Index",y="age",col="class")
dfe
```
