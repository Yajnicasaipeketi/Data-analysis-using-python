```python
#There are 5 sub psychology,nanotechnnology,medicine,ornocology,anthropology.Two students have applied to do masters in any of these 5 fields with the lineplot show in which field they can excel?the cutoff for each field is 70
#st1=[67,45,73,71,57]
#st2=[43,67,68,53,85]
sub=["phsyco","nano","orno","meds","anthro"]
st1=[67,45,73,71,57]
st2=[43,67,68,53,85]
cut=[70,70,70,70,70]
plt.plot(sub,st1,label="st1")
plt.title("")
plt.xlabel("sub")
plt.ylabel("marks")
plt.plot(sub,st2,label="st2")
plt.plot(sub,cut,label="cut")
plt.xlabel("sub")
plt.ylabel("marks")
plt.legend()
```

```python
#There are 5 sub psychology,nanotechnnology,medicine,ornocology,anthropology.Two students have applied to do masters in any of these 5 fields with the lineplot show in which field they can excel?the cutoff for each field is 70
#st1=[67,45,73,71,57]
#st2=[43,67,68,53,85]
sub=["phsyco","nano","orno","meds","anthro"]
st1=[67,45,73,71,57]#student1 marks
st2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student1
#plt.plottype(X,Y,label,color,marker,linewidth,linestyle)
plt.scatter(sub,st1,label="st1",color="c",marker="P")
plt.title("")
#label
plt.xlabel("sub")
plt.ylabel("marks")
#student2
plt.scatter(sub,st2,label="st2",color="m",marker="H")
plt.plot(sub,cut,label="cut",color="k",linestyle =':',linewidth=3)
#linestyle and linewidth only works with plottype=plot
#legend
plt.legend()
```

```python
#An India v/s Aus match is on live.India has given 298 as target for australia.The runs for every 5 overs is given as follows
#India[25,51,84,131,160,189,220,250,267,297]
#Australia have completed only 25 overs
#Australia[15,41,94,110,151,]
#plot overs v/s runs for both the countries with India as a blue line and australia as a yellow line.Give appropriate lables and legend.The marker for India is Flag and australia is plus
# take overs from arange() and create runs as arrays
India=np.array([25,51,84,131,160,189,220,250,267,297])
ovrind=np.arange(5,51,5)
Aus=np.array([15,41,94,110,151])
ovraus=np.arange(5,26,5)
overs=np.array([5,10,15,20,25,30,35,40,45])
plt.plot(ovrind,India,label="India",color="blue",marker=">",linewidth=3)
plt.title(" india v/s aus")
plt.xlabel("Overs")
plt.ylabel("runs")
plt.plot(ovraus,Aus,label="Australia",color="yellow",marker="+",linewidth=3)
plt.xlabel("overs")
plt.ylabel("runs")
plt.grid()
plt.legend()
```
